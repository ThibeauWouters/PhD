# Visualizing sigmoid and relu

```python
import numpy as np
import matplotlib.pyplot as plt

# To get the relevant plot style
import requests
script_url = "https://raw.githubusercontent.com/ThibeauWouters/cheat_sheets/main/myploystyle.py"
exec(requests.get(script_url).text)

plt.rcParams.update({'font.size': 16})
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.where(x < 0, 0, x)
```

```python
bound = 10
x = np.linspace(-bound, bound, 1000)

fs = 26

which = "ReLu"

plt.figure(figsize = (10, 4))
if which == "Sigmoid":
    plt.plot(x, sigmoid(x), color = "blue", lw = 5)
else:
    plt.plot(x, relu(x), color = "blue", lw = 5)
plt.title(which, fontsize = fs)
plt.savefig(f"Plots/{which.lower()}.pdf", bbox_inches = 'tight')
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.show()
```
